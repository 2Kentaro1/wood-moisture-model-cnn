# Colab setup

Private repo `2Kentaro1/wood-moisture-model-cnn` をclone/pullします。

```python
from pathlib import Path
import os

def find_repo_root(start=None):
    start = Path.cwd() if start is None else Path(start).resolve()
    for path in [start, *start.parents]:
        if (path / 'requirements.txt').exists() and (path / 'src').exists():
            return path
    return None

REPO_ROOT = find_repo_root()
if REPO_ROOT is None:
    print('Repo root was not found from:', Path.cwd())
    print('If this is a fresh Colab runtime, set RUN_GIT_SETUP=True in the next cell and run it.')
else:
    os.chdir(REPO_ROOT)
    print('repo root:', Path.cwd())

DATA_DIR = Path('data')
print('train.csv:', (DATA_DIR / 'train.csv').exists())
print('test.csv:', (DATA_DIR / 'test.csv').exists())
```

```python
# Clone/pull this private repo when running in a fresh Colab runtime.
# Set RUN_GIT_SETUP = False only when you are already inside the repository root.
RUN_GIT_SETUP = True

from getpass import getpass
from pathlib import Path
import os
import subprocess

if RUN_GIT_SETUP:
    GITHUB_TOKEN = getpass("GitHub Token: ")
    REPO_URL = "https://github.com/2Kentaro1/wood-moisture-model-cnn.git"
    REPO_NAME = Path(REPO_URL).stem
    AUTH_REPO_URL = REPO_URL.replace("https://", f"https://{GITHUB_TOKEN}@")

    if Path(REPO_NAME).exists():
        subprocess.run(["git", "-C", REPO_NAME, "pull"], check=True)
    else:
        subprocess.run(["git", "clone", AUTH_REPO_URL], check=True)

    os.chdir(REPO_NAME)

Path('data').mkdir(exist_ok=True)
print('repo root:', Path.cwd())
print('Place train.csv and test.csv under:', Path.cwd() / 'data')
```
