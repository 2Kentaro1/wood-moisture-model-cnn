```python
for pattern in ["actual_vs_pred_by_species_*.png", "embedding_pca_*.png", "test_prediction_distribution_*.png", "embedding_feature_correlation_*.png"]:
    matches = sorted(figures_dir.glob(pattern))
    if matches:
        print(matches[-1])
        display(Image(filename=str(matches[-1])))
```
